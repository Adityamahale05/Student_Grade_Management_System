```python
# Initialize the student records dictionary
student_records = { 'Aarav': [97, 76, 61, 91],
    'Vivaan': [80, 49, 41, 90],
    'Aditya': [85, 97, 100, 51],
    'Vihaan': [43, 90, 96, 84],
    'Arjun': [50, 61, 99, 76],
    'Sai': [89, 89, 50, 69],
    'Reyansh': [62, 71, 64, 95],
    'Ayaan': [50, 72, 63, 84],
    'Krishna': [88, 46, 96, 43],
    'Ishaan': [61, 82, 99, 55],
    'Anaya': [97, 93, 94, 74],
    'Siya': [62, 93, 87, 92],
    'Diya': [82, 51, 43, 53],
    'Myra': [87, 59, 75, 66],
    'Ira': [55, 94, 85, 62],
    'Aadhya': [56, 90, 46, 42],
    'Meera': [92, 58, 55, 83],
    'Anvi': [41, 73, 80, 44],
    'Avni': [79, 95, 68, 49],
    'Saanvi': [100, 68, 46, 96],
    'Kabir': [71, 92, 90, 99],
    'Rohan': [48, 58, 54, 82],
    'Aryan': [44, 65, 85, 57],
    'Dev': [48, 67, 56, 64],
    'Om': [67, 60, 69, 59],
    'Tanvi': [84, 54, 87, 49],
    'Sneha': [68, 44, 92, 94],
    'Priya': [65, 64, 46, 50],
    'Riya': [62, 83, 84, 51],
    'Pooja': [91, 58, 61, 99],
    'Kavya': [87, 44, 90, 47],
    'Isha': [74, 53, 77, 62],
    'Nitya': [61, 95, 48, 81],
    'Tanya': [79, 81, 43, 54],
    'Neha': [88, 94, 74, 100],
    'Aisha': [47, 46, 77, 73],
    'Shruti': [68, 53, 99, 52],
    'Muskan': [63, 91, 83, 85],
    'Sanya': [100, 96, 95, 96],
    'Jiya': [97, 66, 60, 51],
    'Manav': [94, 54, 78, 53],
    'Laksh': [44, 66, 96, 57],
    'Yash': [61, 70, 60, 82],
    'Shaurya': [92, 60, 65, 48],
    'Hardik': [78, 90, 73, 61],
    'Nikhil': [65, 46, 88, 53],
    'Rahul': [41, 84, 41, 51],
    'Abhay': [63, 78, 66, 66],
    'Kunal': [82, 65, 46, 53],
    'Siddharth': [91, 54, 76, 71]}

# Function to add a student record
def add_student(name, marks):
    student_records[name] = marks
    print(f"Student '{name}' added successfully.")

# Function to calculate average marks
def calculate_average(marks):
    return sum(marks) / len(marks)

# Function to determine grade
def get_grade(avg):
    if avg >= 90:
        return 'A+'
    elif avg >= 80:
        return 'A'
    elif avg >= 70:
        return 'B'
    elif avg >= 60:
        return 'C'
    else:
        return 'D'
```

```python
# Function to display report of a student
def display_report(name):
    if name in student_records:
        marks = student_records[name]
        avg = calculate_average(marks)
        grade = get_grade(avg)
        print(f"\nReport for {name}:")
        print(f"Marks: {marks}")
        print(f"Average: {avg:.2f}")
        print(f"Grade: {grade}")
    else:
        print("Student not found.")

# Function to show all students
def show_all_students():
    if not student_records:
        print("No student records found.")
    else:
        print("\nAll Student Records:")
        for name in student_records:
            display_report(name)
```

```python
# Main Menu Function
def main():
    while True:
        print("\n==== Student Grade Management ====")
        print("1. Add Student")
        print("2. View Student Report")
        print("3. Show All Students")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter student name: ")
            marks = list(map(int, input("Enter marks separated by space: ").split()))
            add_student(name, marks)

        elif choice == '2':
            name = input("Enter student name to view report: ")
            display_report(name)

        elif choice == '3':
            show_all_students()

        elif choice == '4':
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please try again.")
```
